# Regression
Bei einem Regressionsproblem zielen wir darauf ab, die Ausgabe eines kontinuierlichen Wertes, wie eines Preises oder einer Wahrscheinlichkeit, vorherzusagen. 

![title](img/linreg.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

## Daten generieren

```python
def f(x):
    return 1.5 * x + 4.0
```

```python
def prepare_data():
    N = 100
    x = np.random.uniform(low=-20.0, high=20.0, size=N)
    y = f(x) + np.random.normal(scale=5.0, size=N)
    return x, y
```

```python
x, y = prepare_data()
```

### Daten anzeigen

```python
plt.scatter(x, y, color='g', label='Data')
plt.legend()
plt.grid()
plt.show()
```

## Modell definieren
Dieses Modell dient zum vorhersagen von neuen Punkten bzw. unserem Dataset

```python
def model(x):
    m = 1.0 # slope
    b = 0.0 # intercept

    prediction = m * x + b
    return prediction
```

```python
y_pred = model(x)
```

```python
plt.scatter(x, y, color='g', label='Data')
plt.plot(x, y_pred, color='r', label='Prediction')
plt.legend()
plt.grid()
plt.show()
```

### Wie stark lagen wir daneben?
Hierfür können wir die differenz vom `y` und `y_pred` berechnen.

```python
abs_diff = np.abs(y_pred - y)
mean = np.mean(abs_diff)
mean
```
